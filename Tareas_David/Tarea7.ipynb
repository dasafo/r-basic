## 1- A partir del siguiente código, intenta obtener el gráfico del dibujo llamado ejercicio 1 en el enunciado. Pista: investiga el parámetro fill_between de matplotlib.

n = 256

Y = np.sin(2 * X)

```python
n = 256

X = np.linspace(-np.pi, np.pi, n, endpoint=True)
Y = np.sin(2 * X)

plt.plot(X, Y + 1, color='blue', alpha=1.00)
plt.fill_between(X, 1, Y+1, color='blue', alpha=.15)

plt.plot(X, Y - 1, color='blue', alpha=1.00)
plt.fill_between(X, -1, Y-1, (Y-1) > -1, color='blue', alpha=.15)
plt.fill_between(X, -1, Y-1, (Y-1) < -1, color='red',  alpha=.15)

plt.xticks([])
plt.yticks([])
plt.xlim(-np.pi, np.pi)
plt.ylim(-2.5, 2.5)
plt.show()
```

```python
n = 1024

X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)

plt.scatter(X,Y, c=np.arctan2(Y, X), alpha=.4)

plt.xlim(1,-1 )
plt.ylim(1, -1)
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
n = 12

X = np.arange(n)

Y1 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)

Y2 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)

plt.bar(X, +Y1, facecolor='#9999ff', edgecolor='white')

plt.bar(X, -Y2, facecolor='#ff9999', edgecolor='white')

for x, y in zip(X, Y1): #colocamos el valor encima de las barras superiores

    plt.text(x, y + 0.05, '%.2f' % y, ha='center', va='bottom') 

for x, y in zip(X, Y2): #colocamos el valor encima de las barras inferiores

    plt.text(x , -y - 0.05, '%.2f' % y, ha='center', va='top') 
    
plt.xlim(-1, n) 
plt.ylim(-1.25, +1.25)
plt.xticks([])
plt.yticks([])
```

## 4- A partir del siguiente código, intenta obtener el gráfico del dibujo llamado ejercicio 4 en el enunciado. Pista: tendrás que modificar la Z.

Z = np.random.uniform(0, 1, 20)

plt.pie(Z)

```python
n = 20 #para crear 20 'quesitos'
Z = np.ones(n)
Z[-1] *= 2 #hacemos mas grande el 'quesito' -1(sentido positivo antihorario)
plt.axes([2, 2, 1.5, 1.5])

plt.pie(Z, explode=Z*.06, colors=['%f' % (i/float(n)) for i in range(n)],
        wedgeprops={"linewidth": 0.5, "edgecolor": "black"})

plt.xticks([]) 
plt.yticks([])
plt.show()
```
