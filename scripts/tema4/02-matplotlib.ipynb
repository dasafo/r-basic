# Mltiples grficos en una misma figura

```python
def f(x):
    return np.exp(-x)*np.cos(2*np.pi*x)
```

```python
x1 = np.arange(0, 5.0, 0.1)
x2 = np.arange(0, 5.0, 0.2)
```

```python
plt.figure(1) #para dibujar en una misma figura 2 graficos

plt.subplot(2,1,1) #2 filas, 1 columna, 1 indicamos que aqui ir la primera
plt.plot(x1, f(x1), 'ro', x2, f(x2), 'k')

plt.subplot(2,1,2)
plt.plot(x2, f(x2), 'g--')

plt.show()
```

```python
plt.figure(1)
plt.subplot(2,1,1)
plt.plot([1,2,3])
plt.subplot(2,1,2)
plt.plot([4,5,6])

plt.figure(2)
plt.plot([4,5,6])

plt.figure(1) #podemos volver a la primera y aadir un titulo(por ejemplo)
plt.subplot(2,1,1)
plt.title("Esto es el primer ttulo")
```

# Textos en los grficos

```python
mu = 100
sigma = 20

x = mu + sigma * np.random.randn(10000)
```

```python
#bins son las barras verticales, pachtes cada uno de los conjuntos de datos, hist de histograma
#50 el numero de barras, normed para que aparezcan normalizados a 1
n, bins, patches = plt.hist(x, 50, normed=1, facecolor="g", alpha=0.6) 

plt.xlabel("Cociente Intelectual", fontsize = 14, color = "green")
plt.ylabel("Probabilidad")
plt.title(r"Histograma de CI $N(\mu,\sigma)$") #la r delante le indica que habra latex
plt.text(120, 0.015, r'$\mu = 100,\ \sigma=20$') #colocamos texto
plt.axis([20,180,0, 0.025]) #le indicamos el tamao de ejes que queremos visualizar
plt.grid(True) #para que aada la parrilla de celdas


plt.show()
```

```python
plt.figure(figsize=(8,6), dpi = 90) #con figsize indicamos el tamao del grafico(en px/pulgada)
plt.subplot(1,1,1) #esto no es necesario(esta por defecto) pero tambien se puede poner

x = np.arange(0, 10*np.pi, 0.01)
y = np.cos(x)
plt.plot(x,y, lw = 2.0)

plt.annotate('Mximo Local', xy = (4*np.pi, 1), xytext = (15, 1.5),
            arrowprops = dict(facecolor = "black", shrink = 0.05)) #propiedades de las flechas= a un dicionario
plt.ylim(-2,2)
plt.show()
```

# Cambio de escala

```python
from matplotlib.ticker import NullFormatter
```

```python
mu = 0.5
sd = 0.3
y = mu + sd*np.random.randn(1000)
y = y[(y>0) & (y<1)]
y.sort()
x = np.arange(len(y))
```

```python
plt.figure(figsize=(10, 8))

plt.subplot(2,2,1)
plt.plot(x,y)
plt.yscale("linear")
plt.xscale("linear")
plt.title("Escala Lineal")
plt.grid(True)

plt.subplot(2,2,2)
plt.plot(x,y)
plt.yscale("log")
plt.title("Escala Logartmica")
plt.grid(True)

plt.subplot(2,2,3)
plt.plot(x, y - y.mean())
plt.yscale("symlog", linthreshy=0.01) #logartimo simetrico cada 0.01
plt.title("Escala Log Simtrico")
plt.grid(True)

plt.subplot(2,2,4)
plt.plot(x,y)
plt.yscale("logit") #para escala logistica
plt.title("Escala logstica")
plt.gca().yaxis.set_minor_formatter(NullFormatter())
plt.grid(True)

#movemos ligeramente los graficos
plt.subplots_adjust(top = 0.92, bottom = 0.08, left = 0.1, right = 0.95, hspace = 0.35, wspace = 0.35)

plt.show()
```

# Cambios en los ejes

```python
x = np.linspace(-np.pi, np.pi, 256, endpoint=True) #creamos un vector entre dos valores y 256 cifras, endpoint True para que acabe y empiece en los valores que se le ponen
S, C = np.sin(x), np.cos(x)
```

```python
plt.figure(figsize=(10,8))

plt.plot(x, S, color = "blue", linewidth = 1.2, linestyle = "-", label = "seno")
plt.plot(x, C, color = "green", linewidth = 1.2, linestyle = "-", label = "coseno")
plt.xlim(x.min()*1.1, x.max()*1.1)
plt.ylim(S.min()*1.1, S.max()*1.1)
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], #para indicar las rayas encima del eje x
          [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'+$\pi$'])
plt.yticks(np.linspace(-1,1, 2, endpoint=True),
          ['-1',  '+1'])

ax = plt.gca() #con esto accedemos a la posicion de los ejes y lo almacenamos en ax
ax.spines['right'].set_color('none') #quitamos el eje de la derecha
ax.spines['top'].set_color('none') #quitamos el ehe de arriba
ax.xaxis.set_ticks_position('bottom') #le decimos que el ehe de abajo
ax.spines['bottom'].set_position(('data', 0)) #le decimos que pase su posicion de abajo a (x,y)-->(data,0)
ax.yaxis.set_ticks_position('left') #aqui lo mismo que arriba
ax.spines['left'].set_position(('data', 0))

plt.legend(loc="upper left")


x0 = 2*np.pi/3
#ahora dibujamos la raya punteada ---
plt.plot([x0,x0], [0, np.sin(x0)], color = "blue", linewidth = 2.5, linestyle = "--")
plt.scatter([x0, ], [np.sin(x0), ], 50, color = "blue") #dubujamos el punto 
plt.annotate(r'$\sin(\frac{2\pi}{3}) = \frac{\sqrt{3}}{2}$',
            xy = (x0, np.sin(x0)), xycoords = "data",
            xytext = (+20,+40), textcoords = "offset points", #offset points para que pille +20,+40 por encima del eje y no en coordenadas absolutas
            fontsize = 16, arrowprops = dict(arrowstyle = "->", connectionstyle="arc3,rad=.6"))

plt.plot([x0,x0], [0, np.cos(x0)], color = "green", linewidth = 2.5, linestyle = "--")
plt.scatter([x0, ], [np.cos(x0), ], 50, color = "green")
plt.annotate(r'$\cos(\frac{2\pi}{3}) = -\frac{1}{2}$',
            xy = (x0, np.cos(x0)), xycoords = "data",
            xytext = (-90,-60), textcoords = "offset points", 
            fontsize = 16, arrowprops = dict(arrowstyle = "->", connectionstyle="arc3,rad=.2"))

for label in ax.get_xticklabels() + ax.get_yticklabels(): #hacemos una caja roja semitransparente en los numeros
    label.set_fontsize(16)
    label.set_bbox(dict(facecolor='red', edgecolor='None', alpha = 0.2))

plt.show()
```
