# Cómo crear un data frame

```python
import numpy as np
```

```python
df = pd.DataFrame({
    'A': np.random.rand(10), #generamos 10 numeros aleatorios
    'B': np.random.rand(10),
    'C': np.random.rand(10)
})
```

```python
df
```

```python
df.mean()
```

```python
df.mean(axis='columns') #calculamos la media por filas con 'axis'
```

```python
df2 = pd.DataFrame({
    'key' : ['A', 'B', 'C', 'A', 'B', 'C'],
    'values' : range(6)
},
columns = ['key', 'values']
)
df2
```

```python
df2.groupby('key') #agrupamos los datos de la columna Key(solo eso)
```

```python
df2.groupby('key').mean() #agrupamos los datos de la columna Key y les aplicamos la media
```

```python
df2.groupby('key').median()
```

# Manipulación de data frames

```python
df = pd.DataFrame({
       'key' : ['A', 'B', 'C', 'A', 'B', 'C'],
       'value1': range(6),
       'value2': np.random.randint(0, 100, 6)
})
```

```python
df
```

## Aggregate

```python
df.groupby('key').aggregate(['min', np.median, np.mean, max])
```

```python
df.groupby('key').aggregate({
    'value1': 'min',
    'value2': 'max'
})
```

## Filter

```python
df.groupby('key').std()
```

```python
def filter_func(x):
    return x['value2'].std()>20
```

```python
display('df', "df.groupby('key').std()", "df.groupby('key').filter(filter_func)")
```

```python
df.groupby('key').filter(filter_func)
```

## Transform

```python
df.groupby('key').transform(lambda x: x - x.mean())
```

## Apply

```python
def norm_by_col2(x): #normalizamos la primera columna de datos según la suma de la segunda columna
    x['value1'] /= x['value2'].sum()
    return x
```

```python
df.groupby('key').apply(norm_by_col2)
```

# División de datos por columnas

```python
#Asignamos un 0 a la primera fila, un 1 a la segunda, un 0 a la tercera, un 1 a la cuarta, 
#un 2 la quinta y un 0 a la sexta, y los agrupamos así(en vez de por 'key') y realizamos la suma
L = [0,1,0,1,2,0]
print(df)
df.groupby(L).sum() 
```

```python
df.groupby('key').sum() #aqui pasamos un string que indica el identificador de la columna
```

```python
df.groupby(df['key']).sum() #En cambio aqui le pasamos los valores de la propia columna Key
```

```python
df2 = df.set_index('key') #indicamos que el indice es la columna Key(ya no es 0,1,2,3,4 como se ve arriba)
df2
```

```python
mapping = { #Ahora asignamos solo 2 categorias diferentes en vez de 3
    'A': 'vowel',
    'B': 'consonant',
    'C': 'consonant'
}
```

```python
df2.groupby(mapping).sum() #Ahora vemos que nos suma las columnas B y C como le hemos indicado arriba
```

```python
df2.groupby(str.lower).mean() #pasamos a minusculas las letras
```

```python
df2.groupby([str.lower, mapping]).mean()
```

## Gráficos desde data frames y exportación

```python
import numpy as np
years = [year for year in range(1900, 2020)]
deads = [(y + np.random.uniform(0,100) - 1850) for y in years]
```

```python
df = pd.DataFrame({
    "year": years,
    "deads": deads
})
```

```python
df.head()
```

```python
df.plot(kind = "scatter", x="year", y = "deads", figsize = (10, 8))
```

```python
my_plot = df.plot(x="year", y = "deads", figsize = (10, 8))
```

```python
my_fig = my_plot.get_figure()
```

```python
my_fig.savefig("../../ejercicios/data_frame_plot.png")
```
